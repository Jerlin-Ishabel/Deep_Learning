****Week 1: Exercise 1 - Matrix Multiplication****

```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.dot(A, B)
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
print("Matrix C = A * B:")
print(C)
```

****Week 1: Exercise 2- Eigen Decomposition****

```python
import numpy as np
A = np.array([[4, 2], [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Matrix A:")
print(A)
print("Eigenvalues:")
print(eigenvalues)
print("Eigenvectors:")
print(eigenvectors)

```

****Week 1: Exercise 3- Singular Value Decomposition (SVD)****

```python
import numpy as np
A = np.array([[3, 2, 2], [2, 3, -2]])
U, S, V = np.linalg.svd(A)
print("Matrix A:")
print(A)
print("U matrix:")
print(U)
print("Singular values:")
print(S)
print("V matrix:")
print(V)
```

****Week 1: Exercise 4- Determinant Calculation****

```python
import numpy as np
A = np.array([[6, 1], [2, 3]])
det = np.linalg.det(A)
print("Matrix A:")
print(A)
print("Determinant of A:")
print(det)
```

****Week 1: Exercise 5 - Norm Calculation****

```python
import numpy as np
v = np.array([3, 4])
L1_norm = np.linalg.norm(v, ord=1)
L2_norm = np.linalg.norm(v)
print("Vector v:")
print(v)
print("L1 norm of v:")
print(L1_norm)
print("L2 norm of v:")
print(L2_norm)
```
